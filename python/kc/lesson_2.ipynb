В переменную `n` сохранено целое число — количество секунд. Напишите программу, которая конвертирует заданное количество секунд n в часы, минуты и секунды. Результат сохраните в переменные `hours`, `minutes`, `seconds`.

То есть сначала нужно посчитать, сколько в число n поместилось целых часов, целых минут и сколько после этого осталось секунд.  Рассмотрим пример:

````python
n = 23810

hours = n // 3600 # Считаем количество полных часов в n секундах 
minutes = (n - hours * 3600) // 60 # Считаем количество полных минут, оставшихся после вычитания полных часов
seconds  = (n - hours * 3600) % 60 # Считаем количество секунд, оставшихся после вычитания полных часов и полных минут

print(hours, minutes, seconds)
````

### Задача 10

Переменные и значения можно сравнивать между собой при помощи операций сравнения:

`x < y — x` строго меньше `y`,

`x <= y — x` меньше или равно `y`,

`x > y — x` строго больше `y`,

`x >= y — x` больше или равно `y`,

`x == y — x` равно `y`,

`x != y — x` не равно `y`.

Сравнивать можно и числа, и строки. Результат сравнения — это True или False, в зависимости от того, верное или неверное выражение написано. Со сравнением чисел все понятно:

````python
3 > 2 # True
3 < -5 # False
5 != 6 # пять не равно шести, это правда, поэтому здесь получаем True
10 == 10 #True
````

Сравнение строк устроено чуть хитрее. Начнем с простого и рассмотрим сравнение двух одинаковых строк: `'abc' == 'abc'` вернет нам `True`. Давайте теперь разберемся, почему `'abc' > 'aba'` вернет нам `True`. Сравнение строк в Python происходит лексикографически, то есть посимвольно слева направо. Если символы совпадают, то Python переходит к следующему символу в строке и продолжает сравнение до тех пор, пока не найдет отличие между символами в строке. Если символы в строках различаются, то Python определяет, какая строка "больше" (по алфавиту, в прямом смысле по номеру в алфавите). Если символы до данной позиции совпадают, то строка, у которой следующий символ больше, считается "большей". При этом, если буквы совпадают, то также учитывается размер, маленькая буква больше большой `'a' > 'A'` . Рассмотрим примеры:

````python
str1 = 'hello'
str2 = 'world'
result = str1 < str2 
print(result) 
# В результате будет значение True, так как код символа 'h' меньше, чем код символа 'w'

str1 = 'apple'
str2 = 'banana'
result = str1 > str2 
print(result) 
# В результате будет значение False, так как первый символ 'a' меньше, чем первый символ 'b'

str1 = 'abc'
str2 = 'aba'
result = str1 > str2 
print(result) 
# В результате будет значение True, так как последний символ 'с'больше, чем символ 'a'
````

### Задача 14

Мы уже знаем, что для создания строки достаточно использовать двойные или одинарные кавычки. Все, что находится внутри кавычек, будет восприниматься как строка. 

```
string1 = 'Hello, world!'
string2 = "I'm a Python developer."
```

Если мы выведем string1 и string2 на печать при помощи функции print(), то увидим на экране все символы, которые мы поместили внутри кавычек. Но это не всегда так. В Python, как и в большинстве языков программирования, в строках могут быть использованы специальные символы. Они называются экранированными последовательностями.



Вот некоторые из наиболее часто используемых экранированных последовательностей в Python:

`\n` — перенос строки

`\t` — табуляция  

`\'` — одинарная кавычка

`\"` — двойная кавычка

`\\` — обратный слеш


Так как пробелы и специальные символы учитываются при сравнении. Поэтому очень важно обращать внимание на такие вещи, лишний пробел в конце строки может привести к досадной ошибке в работе программы. 

````python
x = 'слово \\ перенос строки'
print(x)
````
